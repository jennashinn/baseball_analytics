## Formatting the Batting Table

There are a lot of mismatched names between `batting` and `bwar_bat`. I found that the names in each data frame didn't match 100%. 

**Ex:**
- José Abreu and Jose Abreu
- Ronald Acuna and Ronald Acuna Jr. 

This resulted in null values after the merge. That's why I decided to add the ids to `batting` so that I could use that to merge instead of names. 

So, from trial and error, I realized that the easiest way to merge `batting` and `bwar_bat` was to add `mlb_ID` to `batting`. 

After merging, there are 8 null values in the `Name` column. After checking, it seems like these players don't have an MLB ID for whatever reason so I removed them.

```python
from pybaseball import playerid_reverse_lookup

def get_ids(df, war_df): 
    player_ids = df['IDfg']
    name_id = pd.DataFrame(playerid_reverse_lookup(player_ids, key_type='fangraphs'))
    df_ids = df.merge(name_id, left_on='IDfg', right_on='key_fangraphs', how='inner')
    
    merged = pd.merge(war_df, 
                          df_ids, 
                          how='left',
                          left_on=['mlb_ID','year_ID'],
                          right_on=['key_mlbam','Season'])
    
    # Dropping rows with null values in Name
    print("Number of missing names before:", merged['Name'].isna().sum())
    merged_df = merged.dropna(subset=['Name'])
    # Dropping columns with all NaN 
    all_nan_columns= merged_df.columns[merged_df.isna().all()]
    print("Columns that are completely empty:", all_nan_columns)
    
    cleaned_df = merged_df.drop(columns= all_nan_columns)
    
    return cleaned_df
```

```python
merged = get_ids(batting, bwar_bat)
merged.head()
```

## Handling salary values

About 23% of batters in the dataset are missing salaries. Most of the missing values are likely due to clerical input whereby a player's salary for future years may not be filled in, but may be accounted for in an earlier year. 

There are also players who were traded mid-season. The player is accounted for twice for a single season because they played for a different team and the salary may be missing for one of the entries. However, we will treat their salaries as the same for both teams. 

To fill in these missing values, we will assume the following:
- A player's salary is the same across consecutive seasons even if the salary is missing
- Players traded mid-season will have the same salary for that season across both teams

```python
((bat_clean['salary'].isna().sum()/len(bat_clean['salary']))*100).round(2)
```

## Missing Salaries
For the 23% of players with missing salaries, the non-missing salary from the closest season was used. For example, Player X has a salary of $1 million in 2012, but is missing a salary for 2013. His salary in 2013 will be considered $1 million, as well. 

## Adjust for Inflation
To adjust for inflation, the [CPI](https://fred.stlouisfed.org/series/CPIAUCNS) was used. As of Jan 2024, the CPI was updated for annual end of period values from 2000-2022. While the CPI doesn't yet have an annual end of period value for 2023, it does have a monthly value of 307.051 for Nov 2023 which will be used for the 2023 season.

adjusted_dollars = (current_dollars * cpi_new) / cpi_old

```python
# Import CPI
cpi = pd.read_csv('data/CPIAUCNS.csv')
cpi['Year'] = (cpi['DATE'].str.split('-').str.get(0)).astype(int)
cpi = cpi.rename({'CPIAUCNS': 'CPI'}, axis = 'columns')
cpi = cpi[['Year', 'CPI']]
#cpi.loc[len(cpi.index)] = [2023, 307.051]  cpi has since updated 
cpi.tail(2)
```

```python
def clean_salary(df, cpi):
    print("Number of missing salaries before:", df['salary'].isna().sum())
    salary_col = ['salary']
    filled = df.copy()
    # Filling missing values with the player's closest salary using forward and backfill
    filled.loc[:,salary_col] = df.loc[:,salary_col].groupby(filled['name_common']).ffill().bfill()
    # Check
    print("Number of missing salaries after:", filled['salary'].isna().sum())
    
    salary_df = filled.merge(cpi, left_on='Season', right_on='Year', how='left')
    cpi_23 = 307.051
    salary_df['Adjusted'] = round((salary_df['salary'] * cpi_23) / salary_df['CPI'],0)
    
    return salary_df
```

```python
batting_salary = clean_salary(bat_clean, cpi)
batting_salary.head()
```

## Updating Team Names
Since teams have relocated over the years, their Team IDs have changed. We're going to look at those and replace the outdated Team ID with the current Team ID.

```python
def clean_team_names(df): 
    # Dropping one of the Team columns then cleaning
    df.drop('Team', axis = 1, inplace = True)
    df_updated = df.rename(columns={'team_ID': 'Team'})
    # Replacement names
    replacement_names = {'ANA':'LAA', 'FLA':'MIA', 'MON':'WSN','TBD':'TBR'}
    df_updated.loc[:, 'Team'] = df_updated['Team'].replace(replacement_names)
    
    return df_updated
```

```python
cleaned_batting = clean_team_names(batting_salary)
```

```python
def combine_metrics(df): 
    # Take average of similar metrics
    df['O-Swing %'] = df[['O-Swing%', 'O-Swing% (sc)', 'O-Swing% (pi)']].mean(axis=1)
    df['Z-Swing %'] = df[['Z-Swing%', 'Z-Swing% (sc)', 'Z-Swing% (pi)']].mean(axis=1)
    df['Swing %'] = df[['Swing%', 'Swing% (sc)', 'Swing% (pi)']].mean(axis=1)
    df['O-Contact %'] = df[['O-Contact%', 'O-Contact% (sc)', 'O-Contact% (pi)']].mean(axis=1)
    df['Z-Contact %'] = df[['Z-Contact%', 'Z-Contact% (sc)', 'Z-Contact% (pi)']].mean(axis=1)
    df['Contact %'] = df[['Contact%', 'Contact% (sc)', 'Contact% (pi)']].mean(axis=1)
    df['Zone %'] = df[['Zone%', 'Zone% (sc)', 'Zone% (pi)']].mean(axis=1)
    df['Pace_avg'] = df[['Pace', 'Pace (pi)']].mean(axis=1)

    to_drop= ['O-Swing%', 'Z-Swing%', 'Swing%', 'O-Contact%', 'Z-Contact%', 'Contact%', 'Zone%', 
              'O-Swing% (sc)', 'Z-Swing% (sc)', 'Swing% (sc)', 'O-Contact% (sc)', 'Z-Contact% (sc)', 'Contact% (sc)', 'Zone% (sc)', 
              'O-Swing% (pi)', 'Z-Swing% (pi)', 'Swing% (pi)', 'O-Contact% (pi)', 'Z-Contact% (pi)', 'Contact% (pi)', 'Zone% (pi)',
              'Pace', 'Pace (pi)']
    
    data = df.drop(to_drop, axis = 1)
    
    return data
    
```
